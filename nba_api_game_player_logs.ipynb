```python
import pandas as pd
from nba_api.stats.endpoints import playergamelogs, teamgamelogs
import statistics
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 40)
```

```python
gamelogs = playergamelogs.PlayerGameLogs(season_nullable='2019-20')
teamlogs = teamgamelogs.TeamGameLogs(season_nullable='2019-20')
```

```python
playerlogs = gamelogs.get_data_frames()[0]
teamlogs = teamlogs.get_data_frames()[0]
```

```python
playerlogs.groupby('GAME_ID').get_group("0021900001")
```

```python
teams = teamlogs['TEAM_ABBREVIATION'].unique()
for i in teams:
    ids = teamlogs[teamlogs["TEAM_ABBREVIATION"]==i]["GAME_ID"]
    home = teamlogs[(teamlogs["TEAM_ABBREVIATION"]==i) & (teamlogs["MATCHUP"].str.contains("vs"))]
    away = teamlogs[(teamlogs["TEAM_ABBREVIATION"]==i) & (teamlogs["MATCHUP"].str.contains("@"))]
    opponents = teamlogs["GAME_ID"].isin(ids)
    
    opponents = teamlogs[opponents]
    teamopp = opponents[opponents["TEAM_ABBREVIATION"]!=i]
    opphome = teamopp[teamopp["MATCHUP"].str.contains("@")]
    oppaway = teamopp[teamopp["MATCHUP"].str.contains("vs")]
    
    
    print(i + " home",  round(statistics.mean(home["PTS"]),4), round(statistics.stdev(home["PTS"]),4))
    print(i + " away",  round(statistics.mean(away["PTS"]),4), round(statistics.stdev(away["PTS"]),4))
    print(i + " opphome",  round(statistics.mean(opphome["PTS"]),4), round(statistics.stdev(opphome["PTS"]),4))
    print(i + " oppaway",  round(statistics.mean(teamopp["PTS"]),4), round(statistics.stdev(teamopp["PTS"]),4))
    #print(i + " " , round(statistics.mean(teamlogs[teamlogs["TEAM_ABBREVIATION"]==i]["PTS"]),4), round(statistics.stdev(teamlogs[teamlogs["TEAM_ABBREVIATION"]==i]["PTS"]),4))
    #print(i + " opponets: " , round(statistics.mean(teamopp["PTS"]),4),round(statistics.stdev(teamopp["PTS"]),4))
    
```

```python
df3 = teamlogs.merge(right=teamlogs, left_on='GAME_ID', right_on='GAME_ID')
com = df3[df3['TEAM_NAME_x']!=df3['TEAM_NAME_y']]
com
```

```python
#(com["FGA_x"] + .44 * com["FTA_x"] - com["OREB_x"] + com["TOV_x"])/2
#0.96*((com["FGA_x"])+(com["TOV_x"])+0.44*(com["FTA_x"])-(com["OREB_x"]))
com["poss"] = 0.5 * ((com['FGA_x'] + 0.4 * com['FTA_x'] - 1.07 * (com['OREB_x'] / (com['OREB_x'] + com['DREB_y'])) * (com['FGA_x'] - com['FGM_x']) + com['TOV_x']) + (com['FGA_y'] + 0.4 * com['FTA_y'] - 1.07 * (com['OREB_y'] / (com['OREB_y'] + com['DREB_x'])) * (com['FGA_y'] - com['FGM_y']) + com['TOV_y']))
com[["TEAM_NAME_x","poss"]]
```

```python
teams = teamlogs['TEAM_ABBREVIATION'].unique()
for i in teams:
    print(i + ": ", round(statistics.mean(com[com["TEAM_ABBREVIATION_x"] == i]["poss"]),4),"std: ", round(statistics.stdev(com[com["TEAM_ABBREVIATION_x"] == i]["poss"]),4))
```

```python
com["pace"] = 48 * ((2*com["poss"]) / (2 * (com["MIN_x"])))
for i in teams:
    print(i + ": ", round(statistics.mean(com[com["TEAM_ABBREVIATION_x"] == i]["pace"]),4), "std: ", round(statistics.stdev(com[com["TEAM_ABBREVIATION_x"] == i]["pace"]),4))
```
